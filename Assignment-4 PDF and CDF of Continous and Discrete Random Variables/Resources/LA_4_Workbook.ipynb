```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def normal_pdf(x, mu=0.000001, sigma=1.2):
    return (1.0/math.sqrt(2*math.pi*mu*sigma**2)) * math.exp( -1 * (1.0/2.0) * ( (x - mu)/sigma )**2 )
```

```python
a = -10
b = 10
N = 1000
X = np.linspace(a, b, N)
```

```python
Y = np.array([normal_pdf(x) for x in X])
Y = Y / np.sum(Y)
```

```python
plt.plot(X, Y)
plt.fill_between(X, Y, alpha=0.3, color='g')
plt.show()
```
