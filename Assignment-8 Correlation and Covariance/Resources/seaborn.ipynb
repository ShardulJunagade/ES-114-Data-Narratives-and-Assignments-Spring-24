## Selecting all the rows with class name "setosa"

```python
data['species']=='setosa'
```

```python
X = data[data['species']=='setosa']
X
```

```python
X = data[data['species']=='setosa'].iloc[:, :4].to_numpy()
```

```python
X
```

```python
X.shape
```

## Performing EDA (exploratory data analysis) over the set of features in the dataset.

```python
plt.figure(figsize=(18, 4))
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c='r', s=60, alpha=0.4)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.subplot(1, 3, 2)
plt.scatter(X[:, 0], X[:, 2], c='b', s=60, alpha=0.4)
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.subplot(1, 3, 3)
plt.scatter(X[:, 0], X[:, 3], c='g', s=60, alpha=0.4)
plt.xlabel('sepal_length')
plt.ylabel('petal_width')
plt.show()
```
